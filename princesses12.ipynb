```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tokenize import TweetTokenizer
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

#this is sample data
from nltk.corpus import names  

from string import punctuation
```

```python
prs=[]

with open('12dancingprincesses.txt', 'r') as f:
    for line in f:
        cline = line.strip()
        if cline == '': pass  
        else:
            tknls = word_tokenize(cline)
        for token in tknls:
            prs.append(token)
  
```

```python
#remove punctuation 
for token in prs:
    if token in punctuation:
        prs.remove(token)
        
```

```python

prs_stopwords = stopwords.words('english')
prs_stopwords
```

```python
# filler words (stopwords) from the list of tokens.
prs_count = 0
prsnew_words = []  #list to hold new words

for token in prs:
    if token not in prs_stopwords:
        prsnew_words.append(token)
    else: prs_count += 1
```

```python
#Now let's see the new top 10 words in this text.
fd_nw = FreqDist(prsnew_words)
fd_nw.most_common(10)
```
